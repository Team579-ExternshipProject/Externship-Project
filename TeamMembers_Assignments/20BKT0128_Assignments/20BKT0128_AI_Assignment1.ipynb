<h2>TASK-1: Create a pandas dataframe (DataFrame named as 'df') with numpy random values (4 observations and 4 features) </h2>

```python
#Imports
import numpy as np
import pandas as pd
```

```python
#We first generate a random numpy 4X4 array. - 4 observations and 4 features
numpy_random = np.random.randint(low=0, high=50, size=(4,4))
numpy_random
```

```python
#Next, convert this to a pandas dataframe (fn - Feature n)
df = pd.DataFrame(numpy_random, columns=["f1","f2","f3","f4"])
df
```

<h2>TASK-2: Rename the task - 1 'df' dataframe column names to "Random value 1", "Random value 2", "Random value 3" and "Random value 4"</h2>

```python
#Renaming the columns (fn -> Column-n)
df.rename(columns = {'f1':'Random value 1','f2':'Random value 2','f3':'Random value 3','f4':'Random value 4'}, inplace = True)
```

```python
#Printing the data-frame
df
```
